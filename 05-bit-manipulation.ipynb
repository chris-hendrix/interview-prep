## Bit to int conversions

```python
def int2bin(val, bits, curr='', pos=1, sep=''):
    if pos == bits+1: return curr

    exp = bits-pos
    num = 2**exp

    if val < 0:
        val = 2**exp + val
        curr += '1'+sep
    elif pos == 1 and val >= 0:
        val = val
        curr += '0'+sep
    elif val >= num:
        val = val - num
        curr += '1'
    else:
        curr += '0'
    result = int2bin(val, bits, curr, pos+1)
    return result

n, b = -7, 4
print(n, int2bin(n,b))
n, b = -1, 4
print(n, int2bin(n,b))
n, b = 0, 4
print(n, int2bin(n,b))
n, b = 5, 4
print(n, int2bin(n,b))
n, b = 19, 8
print(n, int2bin(n,b))
```

```python
def bin2int(binary, hassign = True):
    bits = len(binary)
    result = 0
    maxbit = 2**(bits-1) 
    if hassign and binary[0] == '1': 
        val = bin2int(binary[1:], False)
        result = val - maxbit
        return result
    elif hassign:
        binary = binary[1:]

    for i in range(len(binary)-1, -1, -1):
        j = len(binary) - 1 - i
        b = int(binary[i])
        result += b * 2**j

    return result

b = '0101'
print(b, bin2int(b))
b = '1111'
print(b, bin2int(b))
b = '1101'
print(b, bin2int(b))
b = '11011010'
res = b2v(b)
print(b, bin2int(b))
print(res, int2bin(res,len(b)))
b = '00010011'
print(b, bin2int(b))
```

## Common bit tasks

```python
# getbit: bit at i from right == 1 (boolean)
def getbit(num, i, bits=8, verbose = False):

    if verbose: 
        print('getbit')
        print('\tnum=\t', int2bin(num, bits), num)
        print('\t1 << i=\t', int2bin(1 << i, bits), 1 <<i)
        print('num & (1 << i)=\t', int2bin(num & (1 << i),bits),num & (1 << i))
        print()
    
    print(int2bin(num, bits))
    return ((num & (1 << i)) != 0)

b = '0101'
num = bin2int(b)
i = 0
print('bin=' + b,'num='+str(num), 'i='+str(i), 'getbit=' + str(getbit(num,i, 4, False)))
i = 1
print('bin=' + b,'num='+str(num), 'i='+str(i), 'getbit=' + str(getbit(num,i, 4)))
i = 2
print('bin=' + b,'num='+str(num), 'i='+str(i), 'getbit=' + str(getbit(num,i, 4)))
```

```python
# setbit: set bit at i from right == 1 (new int)
def setbit(num, i, bits=8, verbose=False):
    if verbose: 
        print('setbit')
        print('\tnum=\t', int2bin(num, bits), num)
        print('\t1 << i=\t', int2bin(1 << i, bits), 1 <<i)
        print('num | (1 << i)=\t', int2bin(num | (1 << i),bits),num | (1 << i))
        print()
    return num | (1 << i)

b = '0101'
num = bin2int(b)
i = 1
print('bin=' + b,'num='+str(num), 'i='+str(i), 'getbit=' + str(setbit(num,i, 4, True)))
```

```python
# clear bit: set bit at i from right == 0 (new int)
def clearbit(num, i, bits=8, verbose=False):
    mask = ~(1 << i)
    return num & mask
```
