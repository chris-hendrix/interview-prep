## Define object

```python
class Node:
  def __init__(self, data = None, next=None): 
    self.data = data
    self.next = next
  
  def __str__(self):
    return "Node data: " + str(self.data)
```

```python
def to_linked_list(data):
    head = Node()
    node = head
    for i in range(len(data)):
        node.data = data[i]
        if i < len(data) - 1 :
            node.next = Node()
        node = node.next
    return head

def to_list(head):
    max = 20
    node = head
    result = []
    count = 0
    while node != None:
        if count > max: break
        result.append(node.data)
        node = node.next
        count += 1
    return result

linked = to_linked_list([1,2,3])
print(linked)
print(to_list(linked))
```

```python
def remdup(head):
    ht = {}
    node = head
    prev = None
    
    while node != None:
        if node.data in ht:
            prev.next = node.next
        else:
            ht[node.data] = 1
            prev = node
        node = node.next
    return head

head = to_linked_list([1,2,3,2,1,0])
result = remdup(head)
print(to_list(result))
```

```python
def klast(head, k):
    node = head
    knode = head
    i = 0
    while node != None:
        if node.next == None:
            if knode == head: knode = None
            return knode
        if i >= k - 1:
            knode = knode.next
        
        node = node.next
        i += 1
    return None

head = to_linked_list([1,2,3,4,5,6])
k = 4
knode = klast(head, k)
print(knode.data)
```

```python
## reverse linked list
def reverse(head):
    node = head
    rnode = None
    rprev = None

    while node != None:
        rnode = Node(node.data)
        rnode.next = rprev
        rprev = rnode
        node = node.next
    return rnode

def ispal(head):
    rev = reverse(head)
    node = head
    rnode = rev

    while node != None:
        if node.data != rnode.data: return False
        node = node.next
        rnode = rnode.next
    return True

head = to_linked_list([1,2,3,4,3,2,1])
result = reverse(head)
print(to_list(result))
print(ispal(head))
```

```python
def intersect(head1, head2):
    node1 = head1

    while node1 != None:
        node2 = head2
        while node2 != None:
            if node1 == node2: return node1
            node2 = node2.next
        node1 = node1.next

    return None 

# define nodes
n1 = Node(1)
n2 = Node(2)
n3 = Node(3)
n4 = Node(4)
n5 = Node(5)
n6 = Node(6)

# first list (n1)
n1.next = n2
n2.next = n5

# second list (n3)
n3.next = n4
n4.next = n5

# intersecting list (n5)
n5.next = n6

result = intersect(n1, n3)
print(result)

```

```python
def circle(head):
    slow = head
    fast = head

    while slow != None:
        slow = slow.next
        fast = fast.next.next
        print('1-slow', slow)
        print('1-fast', fast)
        if slow == fast: break

    slow = head

    while slow != fast:
        print('2-slow', slow)
        print('2-fast', fast)
        slow = slow.next
        fast = fast.next

    return slow

# define nodes
n1 = Node(1)
n2 = Node(2)
n3 = Node(3)
n4 = Node(4)
n5 = Node(5)
n6 = Node(6)
n7 = Node(7)

# first list (n1)
n1.next = n2
n2.next = n3
n3.next = n4
n4.next = n5
n5.next = n6
n6.next = n7
n7.next = n1

result = circle(n1)
#print(to_list(n1))
print('result',result)
```
