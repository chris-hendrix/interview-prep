```python
def is_u(word):
    l = len(word)
    for i in range(l):
        for j in range(i+1,l):
            if word[i]==word[j]: return False

    return True
print(is_u('word'))
```

## 1.3

```python
def url(s):
    start = False
    lst = list(s)
    for i in range(len(lst)-1,-1,-1):
        c = lst[i]
        if not start and c != ' ' : start = True
        if start and c == ' ':
            lst[i] = '%20'
    return ''.join(lst)

print(url('hello world   '))
```

```python
def pal(s):
    s = s.replace(' ','').lower()
    arr = list(s)
    arr.sort()
    cnt = len(arr)
    singles = 0
    i = 0
    while i < len(arr):
        a = arr[i]
        b = None
        if i+1< cnt: b = arr[i+1]

        
        if a == b:
            i += 2

        if singles < 1 and a != b:
            singles += 1
            i += 1
        elif singles > 0 and a != b: 
            return False
    return True

print(pal("atco ctaa"))
```

```python
def edit(a, b):
    if abs(len(a)-len(b))>1: return False
    mis = 0
    j = 0
    for i in range(len(a)):
        if a[i] != b[j]:
            mis += 1
            if mis > 1: return False
            if len(b) == len(a) - 1: j -= 1
            if len(b) == len(a) + 1: i -+ 1
        j += 1
    return True

print(edit('pale','ple'))
print(edit('pale','pales'))
print(edit('pale','bale'))
print(edit('pale','bake'))
```
