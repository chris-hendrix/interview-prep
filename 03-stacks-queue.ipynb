## Stack class definition

```python
class Stack:
    def __init__(self, size):
        self.size = size
        self.items = [None]*size
        self.pointer = -1

    def push(self, item):
        
        # return False if no space left in array
        if self.pointer+1 >= self.size: return False

        # increment pointer
        self.pointer += 1

        # set array value
        self.items[self.pointer] = item

       # return True on success
        return True

    def pop(self):
        item = self.items[self.pointer] 
        self.items[self.pointer] = None
        self.pointer = max(-1, self.pointer-1)
        return item

    def change(self, item):
        self.items[self.pointer] = item

    def peek(self):
        item = self.items[self.pointer]
        return item

    def empty(self):
        return self.items[self.pointer] == None
    
    def __str__(self):
        return str({
            'pointer': self.pointer,
            'items': self.items
        })
```

## Queue Class Definition

```python
class Queue:
    def __init__(self, size = 0):
        self.size = size
        self.items = [None]*size
        self.end = 0
        self.start = 0

    def push(self, item):
        # return false if space is filled
        if self.items[self.end] != None: return False
        # add item
        self.items[self.end] = item
        # increment end pointer
        self.end +=1
        # move end pointer to beginning if outside range
        if self.end >= self.size: self.end = 0
        # return true on success
        return True

    def pop(self):
        # store item to be popped
        item = self.items[self.start]
        # return None if empty
        if item == None: return None
        # remove item
        self.items[self.start] = None
        # increment start
        self.start += 1
        # move start pointer to beginning if outside range
        if self.start >= self.size: self.start = 0
        return item
      
    
    def __str__(self):
        return str({
            'start': self.start,
            'end': self.end,
            'items': self.items
        })
```

## 3.3 Stack of Plates

```python
class SetOfStacks:
    def __init__(self, cap):
        self.cap = cap
        self.stacks = []
        self.add_stack()

    def __str__(self):
        return str({
            'cap': self.cap,
            'count': self.count,
            'stacks': len(self.stacks)
        })

    def add_stack(self):
        self.stack = Stack(self.cap)
        self.stacks.append(self.stack)
        self.count = 0
    
    def remove_stack(self):
        self.stacks.pop()
        self.stack = self.stacks[-1]
        self.count = self.cap
        
    def push(self, item):
        if self.cap == self.count: self.add_stack()
        self.stack.push(item)
        self.count += 1

    def pop(self):
        if len(self.stacks) == 0: return None
        item = self.stack.pop()
        self.count -= 1
        if self.count == 0: self.remove_stack()
        return item

    def pop_at(self, index):
        if len(self.stacks) - 1 < index: return None
        item = self.stacks[index].pop()
        return item
        


```

```python
ss = SetOfStacks(3)
ss.push(1)
ss.push(1)
ss.push(1)
ss.push(1)
print(ss)
ss.pop()
ss.pop()
print(ss)
ss.pop_at(0)
print(ss)
```

## 3.5 Sort Stack

```python
def sort_stack(stack):
    s1 = stack
    s2 = Stack(s1.size)
    val = None

    while not s1.empty():
        item = s1.pop()
        if s2.empty():
            s2.push(item)
            val = None
        elif item > s2.peek():
            val = item
            # empty s2 into s1
            while not s2.empty():
                s1.push(s2.pop())
            # push val into s1
            s1.push(val)
        else:
            s2.push(item)
        
        #print('s1',s1.items)
        #print('s2',s2.items)

    while not s2.empty():
        s1.push(s2.pop())
    
    return s1


s1 = Stack(5)
s1.push(2)
s1.push(1)
s1.push(3)
s1.push(5)
s1.push(4)

print(s1.items)
print(sort_stack(s1).items)
    
```
