## Fibonacci

```python
# dynamic
def fibdyn(n, memo=[]):
    if len(memo) == 0: memo = [0]*(n+1)

    if n==0 or n==1: return n
    
    if memo[n] == 0: memo[n] = fibdyn(n-1, memo) + fibdyn(n-2, memo)
    print(memo)
    return memo[n]
n = 5
print(fibdyn(n))
```

```python
# bottom up
def fibbu(n):
    if n==0 or n==1: return n
    memo = [0,1]
    fib = 0
    for i in range(2, n+1):
        fib = memo[1] + memo[0]
        memo[0] = memo[1]
        memo[1] = fib

    return fib
    
n=5
print(fibbu(n))
```

### 8.1 Triple Step: 

A child is running up a staircase with n steps and can hop either 1 step, 2 steps, or 3
steps at a time. Implement a method to count how many possible ways the child can run up the
stairs.

```python
# recursive
def steps(n):
    if n < 0: return 0
    if n == 1 or n == 0: return 1
    result = steps(n-1) + steps(n-2) + steps(n-3)
    return result
n = 25
print(steps(n))
```

```python
# dynamic
def stepsd(n, memo=[]):
    if len(memo)==0: memo = [0]*(n+1)
    if n < 0: return 0
    if n <= 1: return 1
    if memo[n]==0:
        memo[n] = stepsd(n-1, memo) + stepsd(n-2, memo) + stepsd(n-3, memo)
    return memo[n]
    
n = 25
print(stepsd(n))
```

```python
# bottom up
def stepsb(n):
    if n < 0: return 0
    if n <= 1: return 1
    nm1 = 1
    nm2 = 1
    nm3 = 0

    result = 0
    for i in range(2, n+1):
        result = nm1 + nm2 + nm3
        nm3 = nm2
        nm2 = nm1
        nm1 = result

    return result

n = 25
print(stepsb(n))
```

### 8.2 Robot in a Grid: 

Imagine a robot sitting on the upper left corner of grid with r rows and c columns.
The robot can only move in two directions, right and down, but certain cells are "off limits" such that
the robot cannot step on them. Design an algorithm to find a path for the robot from the top left to
the bottom right.

```python
def gridpath(i, j, r, c, path=[], blocks=[], found=[False]):
    if found[0]: return
    if i == r and j == c: 
        path.append([i, j])
        found[0] = True
        print(path)

    if i > r or j > c:  return
    if [i,j] in blocks: 
        return

    path.append([i,j])
    right = gridpath(i, j+1, r, c, path, blocks, found)
    down = gridpath(i+1, j, r, c, path, blocks, found)

blocks = [
    [2,1],
]
r, c = 2, 2
gridpath(1, 1,  r , c, path = [], blocks=blocks, found=[False])

blocks = [
    [2,2],
    [2,3]
]
r, c = 3, 3
gridpath(1, 1,  r , c, path=[], blocks=blocks, found=[False])
```

### 8.4 Power Set: 

Write a method to return all subsets of a set.

```python
def subsets(arr, index=None):
    if index is None: index = len(arr)-1
    results = []
    if index==-1:
        results.append([])
    else:
        item = arr[index]
        results = subsets(arr, index-1)
        
        for ss in results[:]:
            nss = ss[:]
            nss.append(item)
            results.append(nss)
    return results

print(subsets([1,2,3,4]))
```

### 8.5 Recursive Multiply: 

Write a recursive function to multiply two positive integers without using the
\* operator. You can use addition, subtraction, and bit shifting, but you should minimize the number
of those operations.

```python
def recmult(a,b):
    if b == 1:
        return a
    
    result = a + recmult(a, b-1)
    return result

a, b = 6, 2
print(recmult(a,b))
```

### 8.6 Permutations without Dups: 

Write a method to compute all permutations of a string of unique
characters.

```python
def perm(string):
    
    if len(string) == 1:
        return [string]
    
    results = []
    first = string[0]
    rest = string[1:]
    prev = perm(rest)

    for word in prev:
        for i in range(0, len(word)+1):
            s = word[0:i] + first + word[i:]
            results.append(s)
    return results

string = 'abc'
print(perm(string))
string = 'aab'
print(perm(string))
```

### 8.7 Permutations with Dups: 

Write a method to compute all permutations of a string whose characters
are not necessarily unique. The list of permutations should not have duplicates.

```python
def permdup(string, counts = {}):
    
    if len(string) == 1:
        return [string]
    
    results = []
    first = string[0]
    rest = string[1:]
    prev = permdup(rest, counts)

    for word in prev:
        for i in range(0, len(word)+1):
            s = word[0:i] + first + word[i:]
            if s in counts:
                counts[s] += 1
            else:
                results.append(s)
                counts[s] = 1
    return results

string = 'aabb'
print(permdup(string))
```

### 8.9 Parens: 

Implement an algorithm to print all valid (e.g., properly opened and closed) combinations
of n pairs of parentheses.

EXAMPLE

Input: 3

Output: « () ) ) J «) (», «» () J () ( (», () () ()

```python
def parens(n):
    
    if n == 1:
        return ["()"]

    # array to hold results
    results = []
    # get prev list
    prev = parens(n-1)
    # sandwich, suffix, prefix new parens
    for combo in prev:
        combo1 = "("+combo+")"
        combo2 = combo+"()"
        combo3 = "()"+combo
        results.append(combo1)
        results.append(combo2)
        if combo3 != combo2: results.append(combo3)


    return results

n = 5
print(parens(n))
```

### 8.10 Paint Fill: 

Implement the "paint nil" function that one might see on many image editing programs.
That is, given a screen (represented by a two-dimensional array of colors), a point, and a new color,
nil in the surrounding area until the color changes from the original color.

```python
import random
def get_screen(rows, cols, colors=['0','1','2']):
    results = []
    for i in range(0,rows):
        row = []
        for j in range(0, cols):
            row.append(random.choice(colors))
        results.append(row)
    return results

def printscreen(screen):
    for row in screen:
        print(" ".join(row))

screen = get_screen(5,5)
printscreen(screen)
```

```python
def fill(screen, i, j, color='2', prevcolor = None):
    if prevcolor == None: prevcolor = screen[i][j]
    if screen[i][j] != prevcolor: return
    screen[i][j] = color

    left = screen[i][j-1] if j-1>=0 else None
    right = screen[i][j+1] if j+1<len(screen[0][:]) else None
    down = screen[i-1][j] if i-1>=0 else None
    up = screen[i+1][j] if j+1<len(screen[:][0]) else None

    if left != None and left == prevcolor:
        fill(screen,i,j-1, color, prevcolor)

    if right != None and right == prevcolor:
        fill(screen,i,j+1, color, prevcolor)

    if down != None and down == prevcolor:
        fill(screen,i-1,j, color, prevcolor)

    if up != None and up == prevcolor:
        fill(screen,i-1,j, color, prevcolor)

screen2 = screen[:]
printscreen(screen2)
print()
i, j = 1, 4
fill(screen2,i,j,'1')
printscreen(screen2)
```

### 8.11 Coins: 

Given an infinite number of quarters (25 cents), dimes (10 cents), nickels (5 cents), and
pennies (1 cent), write code to calculate the number of ways of representing n cents.

```python
def coinways(cents, coins=[25,10,5,1], index=0):
    if index >= len(coins)-1: 
        return 1
    coin = coins[index]
    n = 0
    ways = 0
    while n*coin <= cents:
        remaining = cents - n*coin
        ways += coinways(remaining, coins, index+1)
        n += 1

    return ways

n = 12
print(coinways(n))
```

### 8.12 Eight Queens: 

Write an algorithm to print all ways of arranging eight queens on an 8x8 chess board
so that none of them share the same row, column, or diagonal. In this case, "diagonal" means all
diagonals, not just the two that bisect the board.

```python
def get_board(n):
    board = []
    for i in range(0,n):
        row = []
        for j in range(0,n):
            row.append('0')
        board.append(row)
    return board

def printboard(board):
    for row in board:
        print("  ".join(row))

board = get_board(8)
printboard(board)
```

```python
def geti(n, dim): return n // dim

def getj(n, dim): return n % dim

def getn(i, j, dim): return dim*i + j

def setval(n, val, board):
    dim = len(board[:])
    i = geti(n, dim)
    j = getj(n, dim)
    board[i][j] = val

def checkrow(n, board):
    i = geti(n, len(board[:]))
    for val in board[i][:]:
        if val == '1': return False
    return True

def checkcol(n,board):
    j = getj(n, len(board[:]))
    for i in range(0, len(board)):
        if board[i][j] == '1': return False
    return True

def checkdia(n, board):
    dim = len(board)
    iminusj = geti(n, dim) -  getj(n, dim)
    
    for i in range(dim):
        j = (i - iminusj)
        if j >= dim or j<0: continue
        if board[i][j] == '1': return False
    
    return True


def queens(board, nstart=0, ns=[]):
    dim = len(board)
    total = dim*dim
    rowcount = 0

    
    for n in range(nstart, total):
        print(n)
        i = geti(n,dim)
        j = getj(n,dim)

        if (j+1) % dim == 0:

            if rowcount == 0:
                lastn = ns.pop()
                setval(lastn, '0', board)
                # go back one step
                queens(board, nstart=lastn+1, ns=ns)
            rowcount = 0 
        
        if not checkrow(n, board): continue
        if not checkcol(n, board): continue
        if not checkdia(n, board): continue

        ns.append(n)
        setval(n, '1', board)
        printboard(board)
        rowcount += 1
        print(ns)
        
    return board



board = get_board(4)
printboard(queens(board))


    
```
